`````````python
import ast
# we have converted it in to list or dictionary or any data structure
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
`````````

`````````python
movies['genres']=movies['genres'].apply(convert)
`````````

`````````python
movies.head()
`````````

`````````python
movies['keywords']=movies['keywords'].apply(convert)
`````````

`````````python
def convert3(obj): # took three names from cast
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
`````````

`````````python
movies['cast']=movies['cast'].apply(convert)
`````````

`````````python
movies.head()
`````````

`````````python
movies['crew'][0]
`````````

`````````python
def fetch_director(obj): # fetch dictionary haing job as director
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
`````````

`````````python
movies['crew']=movies['crew'].apply(fetch_director)
`````````

`````````python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x]) # helps to convert Science fiction to 
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])   # sciencefiction
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
`````````

`````````python
import nltk
`````````

`````````python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
`````````

`````````python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))# convert love,loves,loved and many more to love
    return " ".join(y)      # helps to convert list into string 
`````````

`````````python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')  # countvectorizer is class , it will maximum occuring 5000 words
`````````

`````````python
vectors= cv.fit_transform(movies['tags']).toarray()
`````````

`````````python
vectors
`````````

`````````python
cv.get_feature_names() #Help to get Most 5000 frequent words as is associated with 5000
# after re running it we dont get similiar words getting new sets of words........
`````````

`````````python
ps.stem('loves')
`````````

`````````python
movies['tags']=movies['tags'].apply(stem)
`````````

`````````python
from sklearn.metrics.pairwise import cosine_similarity   # 0 =Very less similarity , 1= very similiar
# because cos 0 =1 means they are overlapping
# cos 90=0 vectors are prependicular to each other 
`````````

`````````python
similarity = cosine_similarity(vectors)
`````````

`````````python
similarity[0] # it is comparison with itself as well and one movie with rest of the movies 
`````````

`````````python
list(enumerate(similarity[0])) # enumerate helps to get index with distance for ex - (1,0.0714285)
# this is list of tuples
`````````

`````````python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1]) # key allows to let know that on the basis of 
# which it should carry out sorting operation
`````````

`````````python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6] # top five movies , key on which operation will be perforrmed
`````````

`````````python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]  # we get index known as masking
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies.loc[i[0],'title'])  # print title according to id avilable at zeroth index
    
`````````

`````````python
recommend('The Dark Knight Rises')
`````````
